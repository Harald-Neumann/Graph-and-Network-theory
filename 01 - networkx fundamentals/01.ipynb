```python
import networkx as nx
from matplotlib import pyplot as plt # only needed to be imported to set default figure size
plt.rcParams.update({'figure.figsize': (7.5, 7.5)})
%matplotlib inline
```

```python
# Setting seed to get reproducible results
import random
from numpy import random as nprand
seed = hash("Neumann") % 2**32
nprand.seed(seed)
random.seed(seed)
```

```python
G = nx.Graph()

G.add_node("A")
G.add_nodes_from(["B", "C"])

G.add_edge("A", "B")
G.add_edges_from([("B", "C"), ("A", "C")])

nx.draw_networkx(G)
```
