```python
import networkx as nx
from networkx.drawing.nx_pydot import graphviz_layout # impoprt graphviz_layout to draw trees
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.figsize': (10, 10)})


import random
from numpy import random as nprand
seed = hash("Neumann") % 2**32
nprand.seed(seed)
random.seed(seed)
```

```python
complete_graph = nx.complete_graph(10)
nx.draw_shell(complete_graph)
```

```python
gnm_random_graph = nx.gnm_random_graph(8, 15)
nx.draw(gnm_random_graph)
```

Trees

```python
karate_graph = nx.karate_club_graph() # famous network of friendships in a social network see https://en.wikipedia.org/wiki/Zachary%27s_karate_club
plt.figure(figsize=(15, 15))
karate_pos = nx.spring_layout(karate_graph, k=0.75)
nx.draw_networkx(karate_graph, karate_pos)
```

```python
binomial_tree = nx.random_tree(15)
nx.draw(binomial_tree, pos=graphviz_layout(binomial_tree))
```

```python
balanced_tree = nx.balanced_tree(2, 3) # https://en.wikipedia.org/wiki/B-tree
nx.draw(balanced_tree, pos=graphviz_layout(balanced_tree, prog="dot"))
```

```python
full_rary_tree = nx.full_rary_tree(3, 20) # https://en.wikipedia.org/wiki/M-ary_tree
nx.draw(full_rary_tree, pos=graphviz_layout(full_rary_tree, prog="twopi"))
```

```python

single_linked_list = nx.full_rary_tree(1, 16, create_using = nx.DiGraph) # Use create_using to create a directed graph 
nx.draw_circular(single_linked_list)
```

```python
double_linked_list = nx.full_rary_tree(1, 16, create_using = nx.DiGraph) 
double_linked_list.add_edges_from(((u, v) for (v, u) in double_linked_list.edges()))
nx.draw_circular(double_linked_list)
```
